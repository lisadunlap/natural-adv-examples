```python
errs = pd.read_csv("../results/robust_labels.csv")
errs = errs.drop_duplicates(subset="file", keep=False)
blurry_imgs = errs[errs["label"] == "BLURR"]
other_imgs = errs[errs["label"] == "OTHER"]
blurr_test = errs[(errs["label"] == "BLURR") | (errs["label"] == "OTHER")]
```

```python
class_names = errs["class name"].unique() 
classes = list(class_names) + [f"blurry {s}" for s in class_names]
text_inputs = torch.cat([clip.tokenize(f"a photo of a {c}") for c in class_names] + [clip.tokenize(f"a blurry photo of a {c}") for c in class_names]).to(device)
```

```python
def predict(image_input, text_inputs, classes, verbose=True):
    # Calculate features
    with torch.no_grad():
        image_features = model.encode_image(image_input)
        text_features = model.encode_text(text_inputs)

    # Pick the top 5 most similar labels for the image
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)
    similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
    values, indices = similarity[0].topk(5)

    # Print the result
    if verbose:
        print("\nTop predictions:\n")
        for value, index in zip(values, indices):
            print(f"{classes[index]:>16s}: {100 * value.item():.2f}%")
    return classes[indices[0]]
```
